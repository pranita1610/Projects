#### How are in-demand skills trending for Data Analysts?

```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup 
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

#### Aggregating monthly basis

```python
df_DA_US['job_posted_date_month_no'] = df_DA_US['job_posted_date'].dt.month
df_DA_US_skills = df_DA_US.explode('job_skills')

df_DA_US_skills_pivot = df_DA_US_skills.pivot_table(index='job_posted_date_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_DA_US_skills_pivot.loc['Total'] = df_DA_US_skills_pivot.sum()
df_DA_US_skills_pivot = df_DA_US_skills_pivot[df_DA_US_skills_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_US_skills_pivot = df_DA_US_skills_pivot.drop('Total')
df_DA_US_skills_pivot
```

#### Calculate percentage

```python
DA_totals = df_DA_US.groupby(by='job_posted_date_month_no').size()

df_DA_US_percent = df_DA_US_skills_pivot.div(DA_totals/100, axis=0)

df_DA_US_percent
```

#### Change the month no to month name

```python
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_date_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_date_month_no')
df_DA_US_percent
```

#### Find first 5 values

```python
df_plot = df_DA_US_percent.iloc[:, :5]
```

#### Plot

```python
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analyst in the US')
plt.ylabel('Likelihood in Job Postings')
plt.ylim(10, 60)
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])  

plt.tight_layout()
plt.show()
```
